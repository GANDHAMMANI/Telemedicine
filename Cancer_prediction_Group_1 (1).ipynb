Building the Random forest model

```python
# Separate features (X) and target (y)
X = data.drop(columns=['level'])
y = data['level']
```

```python
# Check for any non-numeric columns
non_numeric_columns = X.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python
# Verify the gender encoding
print(X['gender'].unique())  # This should output [0, 1]
```

```python
# Apply Label Encoding to 'gender' column
label_encoder = LabelEncoder()
X['gender'] = label_encoder.fit_transform(X['gender'])
```

```python
# Check for any non-numeric columns in X
non_numeric_columns = X.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to all categorical columns
for col in X.select_dtypes(include=['object']).columns:
    label_encoder = LabelEncoder()
    X[col] = label_encoder.fit_transform(X[col])
```

Splitting the dataset into traning and testing

```python
from sklearn.model_selection import train_test_split

# Re-split the data after encoding
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=100)
rf_model.fit(X_train, y_train)
```

```python

# Feature Importance
importance = rf_model.feature_importances_
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': importance}).sort_values(by='Importance', ascending=False)

# Plot Feature Importance
fig = px.bar(feature_importance,
             x='Feature',
             y='Importance',
             title='Feature Importance (Random Forest)')
fig.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predict on the test set
y_pred = rf_model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")


```

As per the accuracy the RF model is overfitting

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cv_scores = cross_val_score(rf_model, X, y, cv=5)
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean cross-validation score: {cv_scores.mean()}")
```

```python
rf_model = RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42)
rf_model.fit(X_train, y_train)
```

BUilding the KNN model

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


# Train KNN model again
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# Predictions
y_pred_knn = knn_model.predict(X_test)

# Evaluate the model
print(accuracy_score(y_test, y_pred_knn))
```

Classification Report of the KNN

```python
# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))

```

COnfution matrix of the KNN

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_knn)

# Define class labels corresponding to the encoded values
class_labels = ['Low Risk', 'Medium Risk', 'High Risk']

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=class_labels)
disp.plot(cmap="Blues")
disp.ax_.set_title("Confusion Matrix for KNN")
plt.show()
```
